```python
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ignore warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris.data)
y = pd.DataFrame(iris.target)
X.columns = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]
y.columns = ["Name"]
```

```python
df = pd.concat([X, y], axis = 1)
df
```

```python
df.ix[df.Name == 0, ['Name']] = 'Setosa'
df.ix[df.Name == 1, ['Name']] = 'Versicolor'
df.ix[df.Name == 2, ['Name']] = 'Virginica'
```

```python
sns.pairplot(df, hue="Name", diag_kind='kde')
```

```python
plt.figure(figsize=(12,8))
plt.subplot(121)
sns.swarmplot('Name', 'PetalLength', data = df)
plt.subplot(122)
sns.violinplot('Name', 'PetalLength', data = df)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(121)
sns.swarmplot('Name', 'PetalWidth', data = df)
plt.subplot(122)
sns.violinplot('Name', 'PetalWidth', data = df)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(121)
sns.swarmplot('Name', 'SepalLength', data = df)
plt.subplot(122)
sns.violinplot('Name', 'SepalLength', data = df)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(121)
sns.swarmplot('Name', 'SepalWidth', data = df)
plt.subplot(122)
sns.violinplot('Name', 'SepalWidth', data = df)
```
